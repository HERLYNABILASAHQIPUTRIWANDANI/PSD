##  Min max dari setiap kolom


### Minimal setiap kolom
Tabel menampilkan perhitungan nilai minimum dari setiap atribut bunga iris. Pada kelas Iris-setosa, nilai minimum petal_length adalah 1, petal_width sebesar 0,10, sepal_length sebesar 4,30, dan sepal_width sebesar 2,30. Pada kelas Iris-versicolor, nilai minimum petal_length tercatat 3, petal_width sebesar 1, sepal_length sebesar 4,90, dan sepal_width sebesar 2,00. Sementara pada kelas Iris-virginica, nilai minimum petal_length adalah 4,50, petal_width sebesar 1,40, sepal_length sebesar 4,90, dan sepal_width sebesar 2,20.

Secara keseluruhan, nilai minimum dari seluruh data menunjukkan bahwa petal_length terendah adalah 1, petal_width terendah adalah 0,10, sepal_length terendah adalah 4,30, dan sepal_width terendah adalah 2,00.

## Rata - rata setiap kolom

Tabel menampilkan nilai rata-rata dari setiap atribut bunga iris berdasarkan kelasnya. Pada kelas Iris-setosa, rata-rata petal_length adalah 1,46, petal_width sebesar 0,25, sepal_length sebesar 5,01, dan sepal_width sebesar 3,43. Pada kelas Iris-versicolor, rata-rata petal_length tercatat 4,26, petal_width sebesar 1,33, sepal_length sebesar 5,94, dan sepal_width sebesar 2,77. Sementara pada kelas Iris-virginica, rata-rata petal_length adalah 5,55, petal_width sebesar 2,03, sepal_length sebesar 6,59, dan sepal_width sebesar 2,97.

Secara keseluruhan, rata-rata dari seluruh data menunjukkan bahwa petal_length memiliki nilai 3,76, petal_width sebesar 1,20, sepal_length sebesar 5,84, dan sepal_width sebesar 3,06.

## Jumlah setiap kelas ditampilkan dalam grafik batang

Grafik memperlihatkan perbandingan jumlah (sum) dari beberapa atribut bunga iris berdasarkan kelasnya (Iris-setosa, Iris-versicolor, dan Iris-virginica).

Pada kelas Iris-virginica, terlihat nilai tertinggi pada semua atribut, khususnya pada Sum of id yang mencapai lebih dari 6000, sedangkan petal_length, petal_width, sepal_length, dan sepal_width juga lebih besar dibanding kelas lain.

Kelas Iris-versicolor menempati posisi tengah dengan nilai sum yang cukup besar, yaitu sekitar 3775 pada id, serta nilai menengah pada atribut bunga lainnya.

Kelas Iris-setosa memiliki nilai terendah, dengan Sum of id sekitar 1275, dan atribut bunga lain seperti petal_length serta petal_width juga jauh lebih kecil dibanding dua kelas lain.

Secara visual, grafik ini menunjukkan tren bahwa Iris-virginica memiliki ukuran bunga paling besar secara keseluruhan, diikuti oleh Iris-versicolor, dan yang paling kecil adalah Iris-setosa.
