```python
import pandas as pd
import psycopg2
import pymysql

# -----------------------
# Ambil data dari PostgreSQL
# -----------------------
conn_pg = psycopg2.connect(
    host='localhost',
    database='postgres',
    user='postgres',
    password='291104'
)

query_sepal = "SELECT id, class, sepal_length, sepal_width FROM sepal_postgree"
df_sepal = pd.read_sql(query_sepal, conn_pg)
conn_pg.close()

# -----------------------
# Ambil data dari MySQL
# -----------------------
conn_mysql = pymysql.connect(
    host='localhost',
    user='root',
    password='',
    database='petal_mysql'
)

query_petal = "SELECT id, petal_length, petal_width FROM petal"
df_petal = pd.read_sql(query_petal, conn_mysql)
conn_mysql.close()
```

Kode tersebut digunakan untuk mengambil data dari dua jenis basis data yang berbeda, yakni PostgreSQL dan MySQL, dengan bantuan Python dan library pandas.

Pada bagian pertama, kode membuka koneksi ke PostgreSQL menggunakan library psycopg2, lalu menjalankan query untuk mengambil kolom id, class, sepal_length, dan sepal_width dari tabel sepal_postgree. Hasil query disimpan dalam sebuah DataFrame pandas bernama df_sepal agar lebih mudah dianalisis atau diolah. Setelah proses selesai, koneksi ke PostgreSQL ditutup untuk menghindari pemborosan sumber daya.

Selanjutnya, bagian kedua melakukan hal serupa pada MySQL dengan library pymysql. Koneksi dibuat ke database petal_mysql, kemudian query dijalankan untuk mengambil kolom id, petal_length, dan petal_width dari tabel petal. Data yang diambil dimasukkan ke dalam DataFrame df_petal untuk memudahkan manipulasi lebih lanjut di Python. Setelah itu, koneksi ke MySQL juga ditutup.

Dengan cara ini, data dari dua basis data yang berbeda dapat diakses, diintegrasikan, atau dianalisis secara terpisah di dalam Python, sehingga mendukung pengolahan data dari berbagai sumber secara lebih fleksibel.

```python
import pandas as pd
from sqlalchemy import create_engine

# PostgreSQL
engine_pg = create_engine("postgresql+psycopg2://postgres:291104@localhost/postgres")
df_sepal = pd.read_sql("SELECT id, class, sepal_length, sepal_width FROM sepal_postgree", engine_pg)

# MySQL
engine_mysql = create_engine("mysql+pymysql://root:@localhost/petal_mysql")
df_petal = pd.read_sql("SELECT id, petal_length, petal_width FROM petalmysql", engine_mysql)

df_merged = pd.merge(df_sepal, df_petal, on='id', how='inner')

for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    df_merged[col] = df_merged[col].astype(float)
```

Kode tersebut digunakan untuk mengambil sekaligus menggabungkan data dari dua database berbeda, yaitu PostgreSQL dan MySQL, dengan memanfaatkan Python melalui library pandas dan SQLAlchemy.

Pertama, dibuat engine koneksi ke PostgreSQL menggunakan create_engine, lalu dijalankan query untuk mengambil kolom id, class, sepal_length, dan sepal_width dari tabel sepal_postgree. Hasil query ini dimasukkan ke dalam DataFrame pandas bernama df_sepal. Selanjutnya, koneksi serupa dibuat ke MySQL, kemudian diambil data dari tabel petalmysql dengan kolom id, petal_length, dan petal_width, yang kemudian disimpan ke DataFrame df_petal. Pendekatan ini memungkinkan data dari dua sumber berbeda dapat diakses dalam format yang konsisten, yaitu DataFrame.

Setelah data dari kedua database tersedia, dilakukan proses merge antara df_sepal dan df_petal berdasarkan kolom id dengan metode inner join. Hasilnya adalah DataFrame baru bernama df_merged, yang hanya menyimpan baris dengan nilai id yang sama di kedua tabel. Selanjutnya, kolom numerik seperti sepal_length, sepal_width, petal_length, dan petal_width dikonversi ke tipe float agar siap digunakan dalam analisis lebih lanjut. Dengan cara ini, data multi-sumber dari PostgreSQL dan MySQL dapat digabungkan secara efisien untuk mendukung analisis maupun pemodelan di Python.
