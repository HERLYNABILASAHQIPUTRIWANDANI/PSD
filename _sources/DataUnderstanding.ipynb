# Data Understanding

## Import Library

```python
import pandas as pd
from sklearn.datasets import load_iris
```

- import pandas as pd, buat olah data dalam bentuk tabel (DataFrame).

- from sklearn.datasets import load_iris, ambil dataset bunga Iris dari scikit-learn, isinya 150 data dengan 4 fitur (sepal & petal panjang/lebar) dan 3 jenis bunga.

## Mengumpulkan Data Awal

```python
iris = load_iris()

# Membuat DataFrame dari fitur
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Tambahkan kolom target
df['target'] = iris.target

# Info awal dataset
print("Jumlah baris dan kolom:", df.shape)
print("Nama kolom:", df.columns.tolist())
print("\nPreview 5 baris pertama:")
print(df.head())
```

 Memuat dataset Iris menggunakan load_iris(), lalu mengubah data fitur (panjang dan lebar sepal serta petal) menjadi sebuah DataFrame dengan nama kolom sesuai fitur. Setelah itu ditambahkan kolom target yang berisi label kelas bunga (0 = setosa, 1 = versicolor, 2 = virginica).


- Jumlah baris dan kolom dalam dataset (150 baris, 5 kolom).

- Nama-nama kolom pada DataFrame.

- Lima baris pertama sebagai preview isi data.

## Statistik Deskriptif

```python
print("\n Statistik Deskriptif (4 fitur numerik):")
print(df.describe())
```

Hasil statistik deskriptif pada dataset Iris menunjukkan bahwa setiap fitur memiliki karakteristik berbeda:

- Petal width rata-rata 1,19 cm, dengan rentang nilai antara 0,1 cm hingga 2,5 cm.

Target menunjukkan label kelas bunga, bernilai 0, 1, atau 2, yang mewakili tiga jenis Iris: setosa, versicolor, dan virginica.

Secara keseluruhan, fitur petal length dan petal width memiliki rentang nilai yang lebih lebar sehingga lebih informatif untuk membedakan jenis bunga dibandingkan fitur sepal.

## Distribusi Kelas (Target)

```python
print("\n Distribusi Kelas (target):")
print(df['target'].value_counts())

# Mapping angka ke nama bunga
target_mapping = dict(zip(range(3), iris.target_names))
print("\n Mapping target:")
print(target_mapping)
```

Hasil distribusi kelas pada dataset Iris menunjukkan bahwa setiap jenis bunga memiliki jumlah data yang sama, yaitu 50 sampel untuk masing-masing kelas. Dengan demikian, total data berjumlah 150 sampel yang terbagi rata.

Mapping target menunjukkan bahwa angka 0 mewakili Iris-setosa, angka 1 mewakili Iris-versicolor, dan angka 2 mewakili Iris-virginica.

## Verifikasi Kualitas Data

```python
# Cek Missing Value
print("\n Missing Value per Kolom:")
print(df.isnull().sum())

# Cek Tipe Data
print("\n Tipe Data:")
print(df.dtypes)

# Keseimbangan Kelas (proporsi)
print("\n Proporsi tiap kelas:")
print(df['target'].value_counts(normalize=True))

# Range nilai tiap fitur
print("\n Range nilai tiap fitur:")
for col in iris.feature_names:
    print(f"{col} → Min: {df[col].min()} | Max: {df[col].max()}")

# Variansi tiap fitur
print("\n Variansi tiap fitur:")
print(df.var())

# Duplikasi data
print("\n Jumlah data duplikat:")
print(df.duplicated().sum())

# Korelasi antar fitur
print("\n Korelasi antar fitur:")
print(df.corr())



```

- Missing value: tidak ditemukan nilai kosong, sehingga dataset dapat dikatakan bersih.

- Tipe data: seluruh fitur bertipe float64, sedangkan kolom target bertipe int64.

- Proporsi kelas: tiap kelas memiliki proporsi yang sama, yaitu sekitar 33,33%, sehingga distribusinya seimbang.

- Rentang fitur: menunjukkan variasi nilai yang wajar dan realistis pada setiap atribut.

- Variansi: menggambarkan tingkat penyebaran data pada masing-masing kolom.

- Duplikasi data: tidak ada, karena dataset Iris memang tidak mengandung baris ganda.

- Korelasi: terdapat hubungan sangat kuat antara petal length dan petal width (~0,96), yang berarti ukuran petal menjadi indikator utama dalam membedakan jenis bunga.

## Deteksi Outlier pada Dataset Iris

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# 1. Load dataset Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# 2. Deteksi Outlier dengan metode IQR
Q1 = df[iris.feature_names].quantile(0.25)   # Kuartil 1
Q3 = df[iris.feature_names].quantile(0.75)   # Kuartil 3
IQR = Q3 - Q1

outliers = ((df[iris.feature_names] < (Q1 - 1.5 * IQR)) |
            (df[iris.feature_names] > (Q3 + 1.5 * IQR)))

print(" Jumlah outlier per kolom:")
print(outliers.sum())

print("\n Baris yang mengandung outlier (index):")
print(df[outliers.any(axis=1)].index.tolist())

# 3. Visualisasi dengan Boxplot
plt.figure(figsize=(12,6))
for i, col in enumerate(iris.feature_names, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()


```

- Sepal length: distribusinya normal dan tidak menunjukkan adanya outlier yang mencolok.

- Sepal width: ditemukan beberapa outlier, misalnya pada nilai yang sangat kecil sekitar 2,0 cm.

- Petal length: data terlihat cukup bersih dengan sedikit sekali outlier.

- Petal width: terdapat beberapa outlier, namun jumlahnya kecil dan tidak terlalu berpengaruh secara keseluruhan.
