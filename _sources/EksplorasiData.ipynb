# Eksplorasi Data

## Import Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

- pandas → dipakai untuk mengolah dan mengelola data dalam format tabel (DataFrame).

- seaborn → digunakan untuk membuat visualisasi data dengan tampilan yang lebih menarik dan informatif.

- matplotlib.pyplot → menyediakan fungsi dasar untuk membuat berbagai jenis grafik.

- sklearn.datasets.load_iris → berfungsi untuk memanggil dan memuat dataset Iris bawaan dari scikit-learn.

## Load Dataset Iris

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].map(dict(zip(range(3), iris.target_names)))
```

Program ini memuat dataset Iris yang terdiri dari 150 sampel bunga dengan 4 fitur utama dan 3 kelas. Data fitur tersebut kemudian disimpan dalam sebuah DataFrame bernama df. Setelah itu, ditambahkan kolom target yang berisi angka 0, 1, dan 2 sebagai label kelas, serta kolom species yang berisi nama spesies bunga yaitu setosa, versicolor, dan virginica.

Hasil akhirnya adalah sebuah tabel (DataFrame) dengan enam kolom: sepal length, sepal width, petal length, petal width, target, dan species.

## Distribusi Kelas (Target)

```python
plt.figure(figsize=(6,4))
sns.countplot(x='species', data=df, palette="Set2")
plt.title("Distribusi Jumlah Data per Spesies")
plt.show()
```

Program ini digunakan untuk menampilkan jumlah data pada setiap kelas bunga (setosa, versicolor, dan virginica). Hasil perhitungan menunjukkan bahwa dataset Iris bersifat seimbang, yaitu masing-masing spesies memiliki 50 data. Visualisasi ditampilkan dalam bentuk grafik batang yang memperlihatkan jumlah data pada tiap kelas.

## Histogram Distribusi Fitur

```python
df[iris.feature_names].hist(figsize=(10,8), bins=15, color='skyblue')
plt.suptitle("Histogram Distribusi Setiap Fitur")
plt.show()
```

Program ini berfungsi untuk membuat histogram pada setiap fitur dalam dataset, yaitu sepal length, sepal width, petal length, dan petal width. Histogram ini berguna untuk melihat bagaimana sebaran nilai dari masing-masing fitur, apakah cenderung normal, miring (skewed), atau memiliki pola tertentu. Hasil akhirnya berupa empat histogram, masing-masing mewakili satu fitur.

## Heatmap Korelasi

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.iloc[:, :-2].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Antar Fitur")
plt.show()
```


Analisis Korelasi Fitur pada Dataset Iris

Program ini menghitung korelasi antar fitur numerik pada dataset Iris. Korelasi bernilai antara -1 hingga +1:

- +1 menunjukkan hubungan positif sempurna (jika satu naik, yang lain ikut naik).
- -1 menunjukkan hubungan negatif sempurna (jika satu naik, yang lain turun).
- 0 berarti tidak ada hubungan linear.
  Dalam visualisasi heatmap, warna merah menandakan korelasi positif kuat, biru menandakan korelasi negatif, sedangkan putih menunjukkan hampir tidak ada korelasi.

Hasil korelasi antar fitur:

- Sepal Length – Petal Length (0,87) → korelasi positif sangat kuat, semakin panjang sepal biasanya petalnya juga lebih panjang.
- Sepal Length – Petal Width (0,82) → korelasi positif kuat, panjang sepal cenderung seiring dengan lebar petal.
- Petal Length – Petal Width (0,96) → korelasi hampir sempurna, menunjukkan kedua fitur sangat berhubungan dan informasinya mirip (redundan).
- Sepal Width – Petal Length (-0,43) dan Sepal Width – Petal Width (-0,37) → korelasi negatif lemah hingga sedang, artinya semakin lebar sepal biasanya petal justru lebih kecil.
- Sepal Length – Sepal Width (-0,12) → korelasi negatif sangat lemah, hampir tidak ada hubungan.

Kesimpulan:

Petal length dan petal width merupakan fitur dengan hubungan paling kuat, dan keduanya menjadi kunci dalam membedakan spesies bunga.
Sepal width kurang informatif untuk klasifikasi karena korelasinya rendah serta distribusinya tumpang tindih antar kelas.
Sepal length masih relevan dipakai karena memiliki korelasi cukup tinggi dengan fitur petal.
